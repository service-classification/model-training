# Decision Tree

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from xgboost import XGBClassifier
```

```python
df = pd.read_csv('Dataset_test30092021_for_python.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.var()
```

```python
df.rename(columns={'equipment rent':'equipment_rent'}, inplace=True)
```

```python
#разбиваем датасет на тренировочную и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,:-1], df.iloc[:,-1] , test_size=0.33, random_state=42)
```

```python
#Создаем классификатор
clf = DecisionTreeClassifier()
```

```python
#Обучаем классификатор
clf.fit(X_train,y_train)
```

```python
#смотрим скор на тренировочной выборке
clf.score(X_train,y_train)
```

```python
#смотрим скор на тестовой выборке
clf.score(X_test,y_test)
```

```python
#делаем предсказания на тестовой выборке
predictions = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```

```python
#смотрим конфьюжн матрицу 
cf_matrix = confusion_matrix(y_test,predictions)
print(cf_matrix)
```

```python
#смотрим многокласосвую конфьюжн матрицу
multilabel_confusion_matrix(y_test,predictions)
```

# Training the Random Forest model
